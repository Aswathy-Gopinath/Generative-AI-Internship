```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoded_data = df.copy()
```

```python
label_encoder = LabelEncoder()
```

```python
encoded_data['Gender'] = label_encoder.fit_transform(encoded_data['Gender'])
encoded_data[['Gender']].head()
```

```python
one_hot_encoded = pd.get_dummies(encoded_data['Department'],prefix ='Department')
encoded_data = pd.concat([encoded_data,one_hot_encoded],axis = 1)
encoded_data.head()
```

```python
performance_mapping =  {1: 'Low', 2: 'Medium', 3: 'High',5: 'Very High'}
```

```python
encoded_data['Performance_Encoded'] = encoded_data['Performance_Rating'].map(performance_mapping)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
File_path = '/content/preprocessing_practice_dataset.csv'
preprocessing_data = pd.read_csv('/content/preprocessing_practice_dataset.csv')
```

```python
minmax_scaler = MinMaxScaler()
standard_scaler = StandardScaler()

preprocessing_data['Age_MinMax'] = minmax_scaler.fit_transform(preprocessing_data[['Age']])
preprocessing_data['Age_Standard'] = standard_scaler.fit_transform(preprocessing_data[['Age']])

preprocessing_data['Experience_MinMax'] = minmax_scaler.fit_transform(preprocessing_data[['Experience_Years']])
preprocessing_data['Experience_Standard'] = standard_scaler.fit_transform(preprocessing_data[['Experience_Years']])

preprocessing_data['Salary_MinMax'] = minmax_scaler.fit_transform(preprocessing_data[['Monthly_Salary']])
preprocessing_data['Salary_Standard'] = standard_scaler.fit_transform(preprocessing_data[['Monthly_Salary']])
```
